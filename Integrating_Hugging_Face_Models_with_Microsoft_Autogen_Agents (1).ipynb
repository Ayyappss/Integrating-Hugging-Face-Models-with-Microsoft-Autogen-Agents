```python
from transformers import AutoModelForCausalLM, AutoTokenizer

model_name = "gpt2"  # Replace with your chosen model
model = AutoModelForCausalLM.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
def user_proxy(user_input, task_type="chat"):
    return forward_to_assistant(user_input, task_type)
```

```python
def forward_to_assistant(user_input, task_type):
    if task_type == "chat":
        response = handle_chat(user_input)
    elif task_type == "QA":
        response = handle_qa(user_input)
    elif task_type == "code":
        response = handle_code(user_input)
    else:
        response = {"error": "Invalid task type"}
    return response
```

```python
def call_external_function(function_name, *args, **kwargs):
    if function_name == "fetch_data":
        return fetch_data(*args, **kwargs)
    elif function_name == "perform_calculation":
        return perform_calculation(*args, **kwargs)
    # Add more function mappings as needed

def fetch_data(param):
    # Simulate data fetching
    return {"data": f"Fetched data for {param}"}

def perform_calculation(a, b):
    # Simulate a calculation
    return {"result": a + b}
```

```python
from transformers import RagTokenizer, RagRetriever, RagTokenForGeneration

rag_model_name = "facebook/rag-token-nq"
rag_tokenizer = RagTokenizer.from_pretrained(rag_model_name)
retriever = RagRetriever.from_pretrained(rag_model_name, index_name="exact", use_dummy_dataset=True)
rag_model = RagTokenForGeneration.from_pretrained(rag_model_name, retriever=retriever)

def retrieve_and_generate(query):
    inputs = rag_tokenizer(query, return_tensors="pt")
    outputs = rag_model.generate(**inputs)
    response = rag_tokenizer.batch_decode(outputs, skip_special_tokens=True)[0]
    return response
```

```python
def handle_chat(user_input):
    inputs = tokenizer(user_input, return_tensors="pt")
    outputs = model.generate(inputs['input_ids'], max_length=50)
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return {"response": response}
```

```python
def handle_qa(user_input):
    return {"response": retrieve_and_generate(user_input)}
```

```python
def handle_code(user_input):
    # Example of handling code tasks
    return call_external_function("perform_calculation", 2, 3)
```

```python
def process_user_input(user_input, task_type):
    if task_type == "chat":
        response = handle_chat(user_input)
    elif task_type == "QA":
        response = handle_qa(user_input)
    elif task_type == "code":
        response = handle_code(user_input)
    else:
        response = {"error": "Invalid task type"}
    return response
```

```python
# Interactive demo
def interactive_demo():
    print("Welcome to the interactive demo!")
    while True:
        user_input = input("Enter your input: ")
        if user_input.lower() == "exit":
            print("Exiting the demo. Goodbye!")
            break
        task_type = input("Enter task type (chat/QA/code): ")
        if task_type.lower() == "exit":
            print("Exiting the demo. Goodbye!")
            break
        response = process_user_input(user_input, task_type)
        print("Response:", response)

if __name__ == '__main__':
    interactive_demo()
```
